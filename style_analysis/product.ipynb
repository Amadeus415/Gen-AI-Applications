```python
# Search Json structure for products

import json

with open('../style_analysis/amazon.json', 'r') as file:
    data = json.load(file)

print(data)
```

```python
# Display images using image address
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

def display_image_from_url(image_url):
    response = requests.get(image_url)
    img = Image.open(BytesIO(response.content))
    
    plt.figure(figsize=(3, 3))
    plt.imshow(img)
    plt.axis('off')
    plt.show()

# Example usage
url = "https://m.media-amazon.com/images/I/61cVggrR-oL._AC_SY625_.jpg"
display_image_from_url(url)
```

```python
# Display product title, description, price, and image
for product in data['products']:
    print(f"Title: {product['title']}")
    print(f"Description: {product['description']}")
    print(f"Price: {product['price']}")
    print(f"Affiliate Link: {product['affiliate_link']}")
    display_image_from_url(product['image_url'])
    print("\n")

```
